### #1. 모듈 선언 

```python
from libs.Face import FacialLandmarkDetector
from libs.FacialExpression import DataAnalyzer
import libs.Face as Face
import cv2
from libs.FacialExpression import FacialExpressionRecognizer
# FacialExpressionRecognizer 선언하기
facial_expression_recognizer = FacialExpressionRecognizer(kernel='linear', C=1)
# 에크만 피쳐를 사용할 것 (정확도 싸움)
#4시 ~ 4시 15분

# 얼굴 특징점 검출 모듈 선언0
facial_landmark_detector = FacialLandmarkDetector(
        model=Face.FACIAL_LANDMARK_DETECTION_MODEL_DLIB,
        face_detector=Face.FACE_DETECTION_MODEL_OPENCV_DNN)

# 데이터 분석 모듈 선언
data_analyzer = DataAnalyzer()
```

```python
import matplotlib.pyplot as plt


for i in range(1,13):
    # 경로 상의 이미지를 읽어오기
    img_neutral = cv2.imread("./datasets/neutral/n"+str(i)+".jpg")
    img_happiness = cv2.imread("./datasets/happiness/h"+str(i)+".jpg")
    img_anger = cv2.imread("./datasets/anger/anger"+str(i)+".jpg")
    img_sadness = cv2.imread("./datasets/sadness/sad"+str(i)+".jpg")
    img_surprise = cv2.imread("./datasets/surprise/sur"+str(i)+".jpg")
    img_disgust = cv2.imread("./datasets/disgust/dis"+str(i)+".jpg")
    img_fear = cv2.imread("./datasets/fear/fear"+str(i)+".jpg")

    # 추출한 데이터를 따로 저장하기 위한 name 변수 리스트
    
    name_list = ["neutral"+str(i), "happiness"+str(i), "anger"+str(i), "sadness"+str(i), "surprise"+str(i), "disgust"+str(i), "fear"+str(i)]

    
    # 이미지 사이즈를 조정 가로 480, 세로 640
#     img_neutral = cv2.resize(img_neutral, (480, 640))
#     img_happiness = cv2.resize(img_happiness, (480, 640))
#     img_anger = cv2.resize(img_anger, (480, 640))
#     img_sadness = cv2.resize(img_sadness, (480, 640))
#     img_surprise = cv2.resize(img_surprise, (480, 640))
#     img_disgust = cv2.resize(img_disgust, (480, 640))
#     img_fear = cv2.resize(img_fear, (480, 640))
    

    for idx, img in enumerate([img_neutral, img_happiness, img_anger, img_sadness, img_surprise, img_disgust, img_fear]):
        # feed(img): 이미지(BGR 형식)를 입력받아 얼굴 검출 및 얼굴 특징점을 검출한다.
        facial_landmark_detector.feed(img)

        # getIsDetect(): 얼굴 검출 여부를 확인하는 함수. 얼굴이 검출되면 True를, 아니면 False를 반환함.
        if facial_landmark_detector.getIsDetect():

            # getFacialLandmark(): 검출된 얼굴 특징점을 반환하는 함수. 얼굴 영역은 FacialLandmark 객체로 반환된다.
            
            landmarks = facial_landmark_detector.getFacialLandmark()

            # FacialLandmark(x. y) 객체 : x, y 2가지 배열로 이루어져 있다.
            # x : 얼굴 특징점의 x 좌표 리스트
            # y : 얼굴 특징점의 y 좌표 리스트
            # 각각의 리스트는 getX(). getY() 함수로 반환받을 수 있다.
            
            x = landmarks.getX()
            y = landmarks.getY()
            
            #name_list에 저장한 이름에 대한 값을 data_analyzer에 딕셔너리 형식으로 저장
            data_analyzer.appendData(name=name_list[idx], value=[x, -y])

    # 플로팅 결과 출력
    data_analyzer.plotDatas(names=name_list, type='scatter') 
        
```

### #3 데이터 정규화하기
이미지의 해상도 및 이미지 내의 얼굴의 위치, 차지하는 크기 등에 따라 정확한 비교가 어려워짐. 이를 해결하기 위해 정규화(Normalization)를 진행하여 비율적인 값으로 스케일링(Scaling)을 진행

```python
    from libs.FacialExpression import DataAnalyzer
    import numpy as np

    def min_max_normalization(value):
        value = list(value)

        _max = max(value)
        _min = min(value)

        result = []

        for val in value:
            _val = (val - _min) / (_max - _min)
            result.append(_val)

        return np.array(result)
```

가장 보편적인 최대-최소 정규화(min-max normalization)를 진행한 후 결과를 확인

```python
   for i in range(1,13): 
        # Data Analyzer에 입력된 데이터를 다시 불러오기
        landmark_neutral = data_analyzer.getData(name='neutral'+str(i))
        landmark_happiness = data_analyzer.getData(name='happiness'+str(i))
        landmark_anger = data_analyzer.getData(name='anger'+str(i))
        landmark_fear = data_analyzer.getData(name='fear'+str(i))
        landmark_surprise = data_analyzer.getData(name='surprise'+str(i))
        landmark_sadnesss = data_analyzer.getData(name='sadness'+str(i))
        landmark_disgust = data_analyzer.getData(name='disgust'+str(i))

        # Neutral 이미지에서 추출한 landmark를 정규화 
        normalized_landmark_neutral_x = min_max_normalization(landmark_neutral[0])
        normalized_landmark_neutral_y = min_max_normalization(landmark_neutral[1])

        # Happiness 이미지에서 추출한 landmark를 정규화
        normalized_landmark_happiness_x = min_max_normalization(landmark_happiness[0])
        normalized_landmark_happiness_y = min_max_normalization(landmark_happiness[1])
        
        # anger 이미지에서 추출한 landmark를 정규화
        normalized_landmark_anger_x = min_max_normalization(landmark_anger[0])
        normalized_landmark_anger_y = min_max_normalization(landmark_anger[1])
        
        # fear 이미지에서 추출한 landmark를 정규화
        normalized_landmark_fear_x = min_max_normalization(landmark_fear[0])
        normalized_landmark_fear_y = min_max_normalization(landmark_fear[1])
        
        # surprise 이미지에서 추출한 landmark를 정규화
        normalized_landmark_surprise_x = min_max_normalization(landmark_surprise[0])
        normalized_landmark_surprise_y = min_max_normalization(landmark_surprise[1])
        
        # sadnesss 이미지에서 추출한 landmark를 정규화
        normalized_landmark_sadnesss_x = min_max_normalization(landmark_sadnesss[0])
        normalized_landmark_sadnesss_y = min_max_normalization(landmark_sadnesss[1])
        
        # disgust 이미지에서 추출한 landmark를 정규화
        normalized_landmark_disgust_x = min_max_normalization(landmark_disgust[0])
        normalized_landmark_disgust_y = min_max_normalization(landmark_disgust[1])

        # 정규화된 얼굴 특징점 플로팅
        data_analyzer.appendData(name='normalized neutral'+str(i), value=[normalized_landmark_neutral_x, normalized_landmark_neutral_y])
        data_analyzer.appendData(name='normalized happiness'+str(i), value=[normalized_landmark_happiness_x, normalized_landmark_happiness_y])
        data_analyzer.appendData(name='normalized anger'+str(i), value=[normalized_landmark_anger_x, normalized_landmark_anger_y])
        data_analyzer.appendData(name='normalized fear'+str(i), value=[normalized_landmark_fear_x, normalized_landmark_fear_y])
        data_analyzer.appendData(name='normalized surprise'+str(i), value=[normalized_landmark_surprise_x, normalized_landmark_surprise_y])
        data_analyzer.appendData(name='normalized sadnesss'+str(i), value=[normalized_landmark_sadnesss_x, normalized_landmark_sadnesss_y])
        data_analyzer.appendData(name='normalized disgust'+str(i), value=[normalized_landmark_disgust_x, normalized_landmark_disgust_y])
        
        _name_list = ['normalized neutral'+str(i), 'normalized happiness'+str(i), 'normalized anger'+str(i), 'normalized fear'+str(i), 'normalized surprise'+str(i), 'normalized sadnesss'+str(i), 'normalized disgust'+str(i)]
        
        data_analyzer.plotDatas(names=_name_list, type='scatter')
        
```

### #4 기계 학습

기계 학습(Machine Learning) 모델은 데이터의 특징만을 학습한다. 즉 학습 데이터 수가 많을 수록, 추출된 특징이 두 그룹의 차이를 명확히 드러낼 수록 좋은 성능의 모델을 학습시킬 수 있다.

```python
import glob

EMOTIONS = ['anger', 'disgust', 'fear', 'happiness', 'neutral', 'sadness', 'surprise']

dataset_path = "./datasets"

feature_list = []
label_list = []

for emotion_idx, emotion in enumerate(EMOTIONS):
    # 라벨링 
    label = emotion_idx
    # glob으로 읽어올 디렉터리 경로
    glob_path = dataset_path + "/" + EMOTIONS[emotion_idx] + "/*"
    # glob을 통해 디렉토리 안에 모든 파일의 경로를 리스트로 저장
    img_paths = glob.glob(glob_path)
    # 각각의 파일에 접근
    for img_path in img_paths:
        
        # 이미지 불러오기
        img = cv2.imread(img_path)
        
        # 랜드마크 검출
        facial_landmark_detector.feed(img)
        
        if facial_landmark_detector.getIsDetect():
            
            landmark = facial_landmark_detector.getFacialLandmark()
            
            x = landmark.getX()
            y = landmark.getY()
            # 정규화
            x = min_max_normalization(x)
            y = min_max_normalization(y)
            
            # 특징 추출
            features = [y[41]-y[36],y[46]-y[45], y[51],y[39]-y[21], y[42]-y[22], x[48], x[54], -y[44] + y[46], -y[37] + y[41], x[54]-x[48], -y[18], -y[25], y[57] - y[48], y[57] - y[54], -y[21], -y[22]]
            
            # 특징 저장
            feature_list.append(features)
            
            # 라벨링 저장
            label_list.append(label)

facial_expression_recognizer.train(data=feature_list, label=label_list)                      
```
